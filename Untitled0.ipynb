```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")


df.head()
```

```python

df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Average scores by gender
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Average scores by lunch type
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Average scores by test preparation course
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot: Math Score by Gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Score by Gender")
plt.show()

# Histogram: Reading Score Distribution
sns.histplot(df['reading score'], kde=True)
plt.title("Distribution of Reading Scores")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='lunch', y='writing score', data=df)
plt.title("Average Writing Score by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Writing Score")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='test preparation course', y='reading score', data=df)
plt.title("Reading Score by Test Preparation Status")
plt.xlabel("Test Preparation")
plt.ylabel("Reading Score")
plt.show()
sns.barplot(x='test preparation course', y='reading score', data=df)
plt.title("Reading Score by Test Preparation Status")
plt.xlabel("Test Preparation")
plt.ylabel("Reading Score")
plt.show()
```

Insights:
- Girls scored better in reading and writing:

Female students had an average reading score of ~72 and writing score of ~74, while male students averaged ~65 in reading and ~63 in writing.

- Boys slightly outperformed girls in math:

Male students had a higher average math score (69) compared to female students (65).

- Standard lunch is linked to higher scores:

Students with standard lunch scored around 70 in all subjects, whereas those with free/reduced lunch scored 10–12 points lower on average.

- Test preparation boosts scores:

Students who completed the test preparation course scored 5–10 points higher in every subject, especially in reading (74) and writing (74).

- Parental education seems to matter:

Students whose parents had a bachelor’s or master’s degree scored higher (average above 70) compared to those whose parents had only high school education (average around 63–65).
